```python
import pandas as pd
import google.generativeai as genai
import os
genai.configure(api_key='!!!') # put yours
```

```python
df = pd.read_csv('/content/ou.csv')
```

```python
import time
end = []
for i in range(0, len(df),15):
  for j in range(i, i+15):
    abstract = df['Abstract'][j]
    Title = df['Title'][j]
    prompt = f"""
  "The output should be include only answer , without qestions or any additional word "
  "answer the question depend on text and if it explicit in text only"
  The following text:
  {Title}

  {abstract}

  questions about text as follows:

  Is the topic of text about oil and gas industry ?
  Is topic about determining new well location ?
  is text explicit mention  to tool used for evaluation ? (simulation , proxy ,other)
  is text mention to machine learning used in the work or not? (don't involve to Optimization algorithms)
  If machine learning used, mention the name of the algorithm if exist in text (don't menstion to Optimization algorithms)
  is text mention to  optimization algorithms used in the work?
  If optimization algorithms are used, mention the name of the algorithm if exist in text (don't involve to machine learning )
  Write a two-sentences to summarize the approach in very simple way with mention to all previous explicit in text

  The output should be as form:

  Yes or no;
  Yes or no;
  tool name or Non exist;
  Yes or no;
  Machine learning method name or Non exist;
  Yes or no;
  Optimization algorithms name or Non exist;
  two-sentence

  """
    try:
      model = genai.GenerativeModel("gemini-1.5-flash")
      response = model.generate_content(prompt)
      end.append(response.text)
    except Exception as e:
      end.append("non \n non \n non \n non \n non \n non \n non \n non ")
      continue


  print(f"total :{i+15} , 15 processing")
  time.sleep(60)
```

```python
split_end = [x.split(';') for x in end]
```

```python
# prompt: ceak if there is element in split_end have more then 8 print index
for i in range(len(split_end)):
  if len(split_end[i]) > 8:
    print(i,len(split_end[i]))
```

```python
# prompt: write code if bigst then 8 delet 7
for i in range(len(split_end)):
  if len(split_end[i]) > 8:
    split_end[i] ="non \n non \n non \n non \n non \n non \n non \n non ".split('\n')
```

```python
# prompt: delet all \n in side element

for i in range(len(split_end)):
  for j in range(len(split_end[i])):
    split_end[i][j] = split_end[i][j].replace("\n", "")
```

```python
new_data = pd.DataFrame(split_end,columns=['industry', 'well_location', 'tool', 'ml_used', 'ml_algorithm', 'optimization_used', 'optimization_algorithm', 'summary'])
```

```python
new_df = pd.concat([new_df.reset_index(drop=True), new_data], axis=1)
```
