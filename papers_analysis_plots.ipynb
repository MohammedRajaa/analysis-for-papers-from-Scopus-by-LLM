```python
# prompt: write pandas code to read xlsx

import pandas as pd

df = pd.read_excel('/content/jor.xlsx')

df.head()
```

## dwnoload data

```python
# prompt: create data with well_location where is yes

df_yes = df[df['well_location'] == 'Yes']
```

```python
df_yes_yes = df_yes[(df_yes['ml_used'] == 'Yes')]
```

```python
df_yes_yes_after_2019 = df_yes_yes[df_yes_yes['Year'] > 2019]

# Print or use the new DataFrame as needed
df_yes_yes_after_2019
```

```python
# prompt: upload df_yes
from google.colab import files
df_yes_yes_after_2019.to_csv('df_yes_yes_4.csv', encoding = 'utf-8-sig')
files.download('df_yes_yes_4.csv')
```

## stat

```python
# prompt: plot df_yes existing with years in period  stackplot

import matplotlib.pyplot as plt

years = sorted(df_yes['Year'].unique())
data_to_plot = df_yes.groupby('Year').size().reset_index(name='Count')

# Extract the count for each year
counts = data_to_plot['Count'].values

plt.stackplot(years, [counts], labels=['Yes'])
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()
```
